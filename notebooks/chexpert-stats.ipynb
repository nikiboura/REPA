```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/ashery/chexpert/train.csv')

# Κράτα μόνο Frontal
df = df[df['Frontal/Lateral'] == 'Frontal']

print(f"Frontal εικόνες: {len(df):,}")

# 13 labels χωρίς Support Devices
label_cols = [
    'No Finding', 'Enlarged Cardiomediastinum', 'Cardiomegaly',
    'Lung Opacity', 'Lung Lesion', 'Edema', 'Consolidation',
    'Pneumonia', 'Atelectasis', 'Pneumothorax', 'Pleural Effusion',
    'Pleural Other', 'Fracture'
]

# -1 (uncertainty) και NaN γίνονται 0
df_clean = df.copy()
df_clean[label_cols] = df_clean[label_cols].replace(-1, 0).fillna(0)

# Single-label ανά κατηγορία 
print(f"\n{'Label':<30} {'Μόνο αυτό το pathology':>25}")
print("="*57)

total = 0
for label in label_cols:
    other_labels = [l for l in label_cols if l != label]
    mask = (
        (df_clean[label] == 1) &
        (df_clean[other_labels] == 0).all(axis=1)
    )
    count = mask.sum()
    total += count
    print(f"{label:<30} {count:>25,}")

print("="*57)
print(f"{'ΣΥΝΟΛΟ':<30} {total:>25,}")
```
